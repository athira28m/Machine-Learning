## Feature scaling

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss
```

```python
data[['latitude','longitude','minimum_nights','number_of_reviews','reviews_per_month',
     'calculated_host_listings_count','availability_365']] = ss.fit_transform(data[['latitude','longitude','minimum_nights','number_of_reviews','reviews_per_month',
     'calculated_host_listings_count','availability_365']])
data.head()
```

## Feature encoding

```python
data_ohe = pd.get_dummies(data[['neighbourhood','room_type']])
data_ohe.head()
```

```python
data = pd.concat([data,data_ohe],axis=1)
data = data.drop(['neighbourhood','room_type'],axis=1)
data.head()
```

# separate x and y

```python
x = data.drop('price',axis=1)
y = data['price']
```

# split the data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state = 0)
```

# Apply linearregression on the train dataset

```python
from sklearn.linear_model import LinearRegression 
lr = LinearRegression()
lr
```

```python
lr.fit(x_train,y_train)
```

# Perform prediction on the test dataset

```python
y_pred = lr.predict(x_test)
y_pred
```

# Evaluation

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

## Using KNeighborsRegressor

```python
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor()
knr
```

```python
knr.fit(x_train,y_train)
```

```python
y_pred_knr = knr.predict(x_test)
y_pred_knr
```

```python
r2_score(y_test,y_pred_knr)
```

- KNeighborsRegressor is better than LinearRegression for the above data.
